```python
import geopandas as gpd
import fiona
```

```python
# Data from this link: https://www.arcgis.com/home/item.html?id=b537a31aebad4159847255ded34dfc07#overview
# Read File
schools = gpd.read_file("../Schools/schools.geojson", engine = "fiona")
# Create Geo Data Frame with geometry defined from the latitude and longitude provided in the dataset
schools = gpd.GeoDataFrame(
    schools, geometry=gpd.points_from_xy(schools.Longitude, schools.Latitude), crs="EPSG:4326"
)
#schools.head()
schools.to_file("texas_private_schools.geojson")
```
